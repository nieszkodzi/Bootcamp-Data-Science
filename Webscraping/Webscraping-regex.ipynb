```python
import io, os.path, re
import requests, urllib
import numpy as np
import pandas as pd
```

```python
f = requests.get("https://pl.wikipedia.org/wiki/Zygmunt_III_Waza")
f.text
```

```python
wynik = re.findall(pattern = r"<a.*? href=[\"'](.*?)[\"'].*?>(.*?)</a>", string=f.text)
pd.DataFrame(wynik)
```

```python
f = requests.get("https://stooq.pl/q/?s=cdr")
f.text
```

```python
kurs = re.findall(pattern = r"<td.*?>(Kurs).*?<br><b><.*?>(.*?)</span></b.*?</td>", string=f.text)
data = re.findall(pattern = r"<td.*?>(Data)<br><span.*?>(.*?)</span><br><span.*?>(.*?)</span>", string=f.text)
zmiana_bzwzgl = re.findall(pattern = r"<td.*?>(Zmiana)<br>.*?><span.*?>(.*?)</span>", string=f.text)
zmiana_wzgl = re.findall(pattern = r"<td.*?>(Zmiana)<br>.*?><span.*?</span><br><span.*?>(\(.*?\))</span>", string=f.text)
pd.DataFrame({kurs[0][0]:kurs[0][1], data[0][0]:data[0][1],"Godzina":data[0][2], "Zmiana Względna":zmiana_wzgl[0][1],"Zmiana Bezwzględna":zmiana_bzwzgl[0]})
```
