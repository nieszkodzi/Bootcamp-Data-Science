## 2

```python
sql_result = pd.read_sql_query("SELECT DISTINCT type, manufacturer, model, seats, engine FROM planes", con)
sql_result
```

```python
wynik = planes[["type", "manufacturer", "model", "seats", "engine"]].drop_duplicates()
wynik
```

## 3

```python
sql_result = pd.read_sql_query("SELECT COUNT(*), engine FROM planes GROUP BY engine", con)
sql_result
```

```python
wynik = pd.DataFrame(planes.engine.value_counts()).sort_values(by = "engine").reset_index().\
sort_values(by = "index")[["engine", "index"]].reset_index(drop = True)
wynik = wynik.rename({"engine" : "COUNT(*)", "index" : "engine"}, axis=1)
wynik 
```

## 4

```python
sql_result = pd.read_sql_query("SELECT COUNT(*), engine, type FROM planes GROUP BY engine,type", con)
sql_result
```

```python
wynik = pd.DataFrame(planes.groupby(["engine","type"], as_index = False).size()).reset_index().rename({0 : "COUNT(*)"}, axis=1)
wynik.loc[:,::-1]
#na końcu odwracamy kolejnośc kolumn
```

```python
wynik[["COUNT(*)", "engine", "type"]]
```

## 5

```python
sql_result = pd.read_sql_query("SELECT MIN(year), AVG(year), MAX(year),\
engine, manufacturer FROM planes GROUP BY engine, manufacturer", con)
sql_result
```

```python
wynik = planes.groupby(["engine", "manufacturer"], as_index = False)["year"].agg({"min","max","mean"}).reset_index()
wynik = wynik.rename({"min" : "MIN(year)", "max" : "MAX(year)", "mean" : "AVG(year)" }, axis=1)
wynik[["MIN(year)", "AVG(year)", "MAX(year)", "engine", "manufacturer" ]]
```

## 12

```python
sql_result = pd.read_sql_query('SELECT manufacturer, COUNT(*) FROM planes GROUP BY manufacturer HAVING COUNT(*) > 10', con)
sql_result
```

```python
wynik = planes.groupby('manufacturer').size()
wynik[wynik > 10].reset_index().rename({0 : "COUNT(*)"}, axis=1)
```

## 13

```python
sql_result = pd.read_sql_query('SELECT manufacturer,COUNT(*)\
FROM planes WHERE seats > 200 GROUP BY manufacturer HAVING COUNT(*) > 10', con)
sql_result
```

```python
wynik = planes[planes.seats > 200].groupby('manufacturer').size()
wynik[wynik > 10].reset_index().rename({0 : "COUNT(*)"}, axis=1)
```

## 14

```python
sql_result = pd.read_sql_query('SELECT manufacturer, COUNT(*) AS howmany \
FROM planes GROUP BY manufacturer ORDER BY howmany', con)
sql_result
```

```python
planes.groupby('manufacturer').size().sort_values().reset_index().rename({0 : "howmany"}, axis=1)
```

## 15

```python
sql_result = pd.read_sql_query('SELECT manufacturer, COUNT(*) AS howmany FROM \
planes GROUP BY manufacturer ORDER BY howmany DESC LIMIT 10', con)
sql_result
```

```python
planes.groupby('manufacturer').size().sort_values(ascending = False).head(10).reset_index().rename({0 : "howmany"}, axis=1)
```

```python
A = airports[0:10]
B = airports[5:15]
```

## SELECT * FROM A UNION ALL SELECT * FROM B

```python
pd.concat([A,B]).shape
```

## SELECT * FROM A UNION SELECT * FROM B

```python
pd.concat([A,B]).drop_duplicates()
```

## SELECT * FROM A INTERSECT SELECT * FROM B

```python
A[A.isin(B)].dropna()
```

```python
pd.merge(A,B)
```

## SELECT * FROM A EXCEPT SELECT * FROM B

```python
A[~A.isin(B)].dropna()
```
