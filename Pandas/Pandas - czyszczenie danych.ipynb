## Maile studentów

```python
oceny["Maile"] = oceny.Nazwisko + oceny.Imię.str[0:1] + "@student.mini.pw.edu.pl"
oceny.Maile = oceny.Maile.str.lower()

ltrPL = "ŻÓŁĆĘŚĄŹŃżółćęśąźń"
ltrnoPL = "ZOLCESAZNzolcesazn"
trantab = str.maketrans(ltrPL, ltrnoPL)
trantab
```

```python
oceny.Maile = oceny.Maile.str.translate(trantab)
```

```python
import re

pattern = r"Z(\d+)\((\d+)\+(\d+)\)"
re.findall(pattern, "Z12(43+56) Z2(34+12))")
```

```python
oceny2 = oceny.copy()

oceny2 = oceny2.rename(columns = {"Z13(5)":"Z13(4+1)"})

del oceny2['Z15(5)']

for col in oceny2.columns:
    if '+' in col:
        name_extr = re.findall(pattern, col)
        col0_name = "Z"+name_extr[0][0]+"("+name_extr[0][1]+")"
        col1_name = "Z"+name_extr[0][0]+"("+name_extr[0][2]+")"
        oceny2[col0_name] = oceny2[col].str.split("+", expand = True).loc[:,0]
        oceny2[col1_name] = oceny2[col].str.split("+", expand = True).loc[:,1]
        del oceny2[col]
        
oceny2
```

## Oceny studentów

```python
oceny2 = oceny2.replace(['NA', r'NS.+', 'NB'], '0', regex=True)
oceny2 = oceny2.replace(['US','ZW'], np.NaN, regex=True)

oceny2
```

```python
oceny3 = oceny2.copy()
oceny3 = pd.concat([oceny2.iloc[:,0:3],oceny2.iloc[:,3:].astype(float)], axis=1)

for columnIndex in range(3, 30):
    for rowIndex in range(oceny3.shape[0]):
        if np.isnan(oceny3.iloc[rowIndex, columnIndex]):
            oceny3.iloc[rowIndex, columnIndex] = \
                np.mean(np.r_[oceny3.iloc[:, columnIndex].mean(),\
                              oceny3.iloc[rowIndex, 3:31].mean()])  
oceny3
```

```python
wynik = oceny3.columns.str.extract("\((\d+)\)")[0].astype(float).sum()
oceny2["Procent"] = oceny3.agg("sum",axis=1)/wynik*100

oceny2["Oceny"] = pd.cut(oceny2.Procent, [0, 50, 60, 70, 80, 90, 100], labels=[2.0, 3.0, 3.5,4.0,4.5,5.0])
oceny2
```
