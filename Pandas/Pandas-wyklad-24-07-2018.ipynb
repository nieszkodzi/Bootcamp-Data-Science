## Tabele przestawne

```python
pd.pivot_table(tips[["smoker", "total_bill", "tip"]],
               index=["smoker"],
               aggfunc="mean"
              )
```

```python
pd.pivot_table(tips[["smoker", "day", "total_bill", "tip"]],
               index=["smoker"],
               columns = ["day"],
               aggfunc="mean"
              )
```

```python
pd.pivot_table(tips[["smoker", "day", "total_bill", "tip"]],
               index=["smoker","day"],
               aggfunc="mean"
              )
```

```python
tips2 = tips.copy()
```

```python
tips2["wielkosc"] = pd.cut(tips2.total_bill,\
                                np.r_[0, 10, 20, 30, 60],\
                                labels=["maly obiadek", "sredni", "dalej sredni", "duzy"])
```

```python
pd.pivot_table(tips2[["wielkosc", "tip"]],
               index=["wielkosc"],
               aggfunc="mean"
              )
```
