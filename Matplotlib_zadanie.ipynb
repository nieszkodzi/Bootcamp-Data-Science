```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
import datetime

matplotlib.__version__
```

```python
plt.figure(figsize=(10,10))
t = np.linspace(0,2*np.pi)
x = 4*np.sin(t)
y = 4*np.cos(t)
r = 4

teraz = datetime.datetime.now()
hour = teraz.hour
minute = teraz.minute
sec = teraz.second

plt.plot([0,0.8*r*np.sin(sec*np.pi/30)], [0,0.8*r*np.cos(sec*np.pi/30)], color="k")
plt.plot([0,0.7*r*np.sin(minute*np.pi/30)], [0,0.7*r*np.cos(minute*np.pi/30)], color="k")
plt.plot([0,0.5*r*np.sin(hour*np.pi/6)], [0,0.5*r*np.cos(hour*np.pi/6)], color="k")

plt.fill(x,y,facecolor="y",edgecolor="r", linewidth=7)
plt.axis("equal")
plt.plot(0,0,"ko")
plt.text(-0.25,3.5,"12",fontsize=30)
plt.text(-0.25,-3.5," 6",fontsize=30)
plt.text(-3.5,0,"9",fontsize=30)
plt.text(3.5,0,"3",fontsize=30)

plt.show()
```

```python
flights=sns.load_dataset("flights")
sredniPasazerowie = flights["passengers"].groupby(flights["year"]).agg(np.mean)
x = sredniPasazerowie.index
y = sredniPasazerowie.values
plt.plot(x,y,"bp--",label = "Średnia miesięczna liczba pasażerów")
plt.legend()
plt.show()
```

```python
tips = sns.load_dataset("tips")
smoker = tips.smoker.value_counts()

plt.pie(smoker.values, 
        textprops={"color" : "white"},
        explode = (0.2,0.3), 
        startangle = 66,
        shadow = True,
        labels=smoker.index,colors=["r","b"])
leg = plt.legend(title="Legenda", frameon=True)
plt.axis("equal")
plt.show()
```

```python
fig, ax = plt.subplots() 
wykres = ax.bar(smoker.index, smoker.values, color=["r","b"], yerr = 10)
ax.set_xticks(rotations = 15)

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 0.8*height,
                '%d' % int(height),
                ha='center', va='bottom')
autolabel(wykres)
plt.show()
```
